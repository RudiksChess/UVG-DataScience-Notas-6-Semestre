## Descendencia por gradiente:  Ejemplo

Buscar el mínimo de 5x² + 3x -4

Regla de actualización:  x(i+1) = x(i) - eta*f´(x(i))

Probemos con valores de **eta** = 0.01, 0.1, 0.001, 0.2

```python
x = [4]
f_prima = [43]
eta = 0.01
print(f"{0}  {x[0]}   {f_prima[0]}")
```

```python
for i in range(1, 1000):
    
    x.append(x[i - 1] - eta * f_prima[i - 1])
    f_prima.append(10 * x[i] + 3)
    print(f"{i}  {x[i]}   {f_prima[i]}")
    
```

Vemos que converge a x = -0.3 alrededor de la 80 y pico iteración

La velocidad de minimización depende de **eta**, probemos con **eta** = 0.1, 0.001 y 0.2
